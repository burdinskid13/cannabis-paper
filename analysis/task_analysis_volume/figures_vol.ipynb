```python
import glob 
import os
from PIL import Image
from PIL import ImageDraw
from PIL import ImageOps
from PIL import ImageFont
```

```python
def create_vol_figure_contrast(group,ses,task,contrast):


    axial_HC_baseline_path = glob.glob(f'../../../derivatives/task_analysis_volume/visualization/raw_figures/group-HC_ses-baseline_task-{task}_contrast-{contrast}_threshold-*_n-*_display-axial.png')[0]
    axial_HC_baseline=Image.open(axial_HC_baseline_path,mode='r')

    w,h=axial_HC_baseline.size
    area_no_cmap = (0, 20, w-83, h-25)

    HCb = axial_HC_baseline.crop(area_no_cmap) #crop just the part of the image that you want


    axial_MM_baseline_path = glob.glob(f'../../../derivatives/task_analysis_volume/visualization/raw_figures/group-MM_ses-baseline_task-{task}_contrast-{contrast}_threshold-*_n-*_display-axial.png')[0]
    axial_MM_baseline=Image.open(axial_MM_baseline_path,mode='r')
    MMb = axial_MM_baseline.crop(area_no_cmap) #crop just the part of the image that you want


    axial_MM_1year_path = glob.glob(f'../../../derivatives/task_analysis_volume/visualization/raw_figures/group-MM_ses-1year_task-{task}_contrast-{contrast}_threshold-*_n-*_display-axial.png')[0]
    axial_MM_1year=Image.open(axial_MM_1year_path,mode='r')

    area_cmap = (0, 20, w, h-25)

    MM1 = axial_MM_1year.crop(area_cmap) #crop just the part of the image that you want


    w_no_cmap,h_no_cmap = HCb.size
    w_cmap,h_cmap = MM1.size

    new_im = Image.new('RGB', ( (w_no_cmap*2+w_cmap) , h_no_cmap) ,(255, 255, 255, 1)) #make a new image
    #paste in the cut out images
    new_im.paste(HCb,(0,0)) 
    new_im.paste(MMb,(w_no_cmap,0))
    new_im.paste(MM1,(w_no_cmap*2,0))
    
    #create paths to output dir if not exist
    derivatives_path = '../../../derivatives'
    nilearn_output_path = os.path.join(derivatives_path, 'task_analysis_volume','visualization','final_nifti_figures')
    if not os.path.isdir(nilearn_output_path):
        os.makedirs (nilearn_output_path)

    HCb_threshold = axial_HC_baseline_path.split('threshold-')[1].split('_')[0]
    HCb_part_count = axial_HC_baseline_path.split('_n-')[1].split('_')[0]

    MMb_threshold = axial_MM_baseline_path.split('threshold-')[1].split('_')[0]
    MMb_part_count = axial_MM_baseline_path.split('_n-')[1].split('_')[0]

    MM1_threshold = axial_MM_1year_path.split('threshold-')[1].split('_')[0]
    MM1_part_count = axial_MM_1year_path.split('_n-')[1].split('_')[0]

    new_im.save(f'../../../derivatives/task_analysis_volume/visualization/final_nifti_figures/task-{task}_contrast-{contrast}_HCb_threshold-{HCb_threshold}_HCb_n-{HCb_part_count}_MMb_threshold-{MMb_threshold}_MMb_n-{MMb_part_count}_MM1_threshold-{MM1_threshold}_MM1_n-{MM1_part_count}_nifti_figure.png')

    return w_no_cmap
```

```python
def create_vol_figure_all(task,contrasts,w_no_cmap):
    png_paths = glob.glob(f'../../../derivatives/task_analysis_volume/visualization/final_nifti_figures/task-{task}_contrast-*_nifti_figure.png')
    
    font = ImageFont.truetype("/usr/share/fonts/Liberation/LiberationMono-Bold.ttf",30)

    if task == 'nback':
        Twoback_Zeroback=Image.open(png_paths[0],mode='r')
        w,h=Twoback_Zeroback.size
        new_im = Image.new('RGB', ( (w) , h) ,(255, 255, 255, 1)) #make a new image
        new_im.paste(Twoback_Zeroback,(0,0))
        
        draw = ImageDraw.Draw(new_im)
        
        draw.text((10,10),"HC baseline",fill=(255,255,255),font=font)
        draw.text((10+w_no_cmap,10),"MCC baseline",fill=(255,255,255),font=font)
        draw.text((10+2*w_no_cmap,10),"MCC one-year",fill=(255,255,255),font=font)

        display(new_im)
        
        new_im.save(f'../../../derivatives/task_analysis_volume/visualization/final_nifti_figures/task-{task}_contrast1-{contrasts[0]}_nifti_figure.png')
    
    if task == 'mid':
        Loss_AvoidLoss=Image.open(png_paths[0],mode='r')
        Win_NoWin=Image.open(png_paths[1],mode='r')
        w,h=Loss_AvoidLoss.size
        
        new_im = Image.new('RGB', ( (w) , h*2) ,(255, 255, 255, 1)) #make a new image
        #paste in the cut out images
        new_im.paste(Win_NoWin,(0,0))
        new_im.paste(Loss_AvoidLoss,(0,h))
        
        
        draw = ImageDraw.Draw(new_im)
        
        draw.text((10,10),"HC baseline",fill=(255,255,255),font=font)
        draw.text((10+w_no_cmap,10),"MCC baseline",fill=(255,255,255),font=font)
        draw.text((10+2*w_no_cmap,10),"MCC one-year",fill=(255,255,255),font=font)

        display(new_im)
        
        
        new_im.save(f'../../../derivatives/task_analysis_volume/visualization/final_nifti_figures/task-{task}_contrast1-{contrasts[0]}_contrast2-{contrasts[1]}_nifti_figure.png')
    
    if task == 'sst':
        Succ_Go=Image.open(png_paths[0],mode='r')
        Unsecc_Go=Image.open(png_paths[1],mode='r')
        Unsecc_Succ=Image.open(png_paths[2],mode='r')
        w,h=Succ_Go.size
        
        new_im = Image.new('RGB', ( (w) , h*3) ,(255, 255, 255, 1)) #make a new image
        #paste in the cut out images
        new_im.paste(Succ_Go,(0,0))
        new_im.paste(Unsecc_Go,(0,h))
        new_im.paste(Unsecc_Succ,(0,h*2))
        
        draw = ImageDraw.Draw(new_im)
        
        draw.text((10,10),"HC baseline",fill=(255,255,255),font=font)
        draw.text((10+w_no_cmap,10),"MCC baseline",fill=(255,255,255),font=font)
        draw.text((10+2*w_no_cmap,10),"MCC one-year",fill=(255,255,255),font=font)

        display(new_im)
        
        new_im.save(f'../../../derivatives/task_analysis_volume/visualization/final_nifti_figures/task-{task}_contrast1-{contrasts[0]}_contrast2-{contrasts[1]}_contrast3-{contrasts[2]}_nifti_figure.png')

    return
```

```python
groups=['HC','MM']
sessions=['baseline','1year']
task='sst'


if task == 'mid':
    contrasts=['HiWin+LoWin-HiNoWin-LoNoWin','HiLoss+LoLoss-AvoidHiLoss-AvoidLoLoss']

elif task == 'sst':
    contrasts=['SuccStop-Go','UnsuccStop-Go','UnsuccStop-SuccStop']

elif task == 'nback':
    contrasts=['twoback-zeroback']



for group in groups:
    for ses in sessions:
        for contrast in contrasts:
            w_no_cmap = create_vol_figure_contrast(group,ses,task,contrast)

create_vol_figure_all(task,contrasts,w_no_cmap)
```
